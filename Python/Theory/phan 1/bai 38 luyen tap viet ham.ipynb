```python
from math import *

def ham1(n) :
    if n < 2 : return 0
    for i in range(2, isqrt(n) + 1) :
        if n % i == 0 : 
            return 0
    return 1

def ham2(n) :
    sum = 0
    while n :
        sum += n % 10
        n //= 10
    return sum

def ham3(n) :
    sum = 0
    while n :
        sum += n % 10 if n % 10 % 2 == 0 else 0
        n //= 10
    return sum

def ham4(n) :
    sum = 0
    while n : 
        r = n % 10
        if ham1(r) : sum += r
        n //= 10
    return sum

def ham5(n) :
    rev = 0
    while n :
        rev = rev * 10 + n % 10
        n //= 10
    return rev

def ham6(n) :
    cnt = 0
    for i in range(2, isqrt(n) + 1) :
        if n % i == 0 :
            cnt += 1
            while n % i == 0 :
                n //= i
    if n != 1 : cnt += 1
    return cnt

def ham7(n) :
    mx = 0
    for i in range(2, isqrt(n) + 1) :
        if n % i == 0 :
            mx = i
            while n % i == 0 :
                n //= i
    if n != 1 :
        mx = n
    return mx

def ham8(n) :
    while n : 
        if n % 10 == 6 : return 1
        n //= 10
    return 0

def ham9(n) :
    if ham2(n)  % 8 == 0 : return 1
    return 0

def ham10(n) :
    sum = 0
    while n :
        sum += factorial(n % 10)
        n //= 10
    return sum

def ham11(n) :
    r = n % 10
    while n :
        if r != n % 10 :
            return 0
        n //= 10
    return 1

def ham12(n) :
    r = n % 10
    while n :
        if r < n % 10 :
            return 0
        n //= 10
    return 1

def ham13(n) :
    tmp = n
    cnt = 0
    while n :
        cnt += 1
        n //= 10

    sum = 0
    while tmp :
        sum += (tmp % 10) ** cnt
        tmp //= 10
    
    return sum

if __name__ == '__main__' :
    n = int(input())
    print(ham1(n), ham2(n), ham3(n), ham4(n), ham5(n), 
          ham6(n), ham7(n), ham8(n), ham9(n), ham10(n), ham11(n), ham12(n), ham13(n), sep = '\n')
```
