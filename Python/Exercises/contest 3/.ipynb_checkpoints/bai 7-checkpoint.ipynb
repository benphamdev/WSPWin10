```python
from math import *

def gcd(a, b) :
    while b :
        a, b = b, a % b
    return a

def lcm(a, b) :
    return a * gcd(a, b) // b

if __name__  == '__main__' :
    a, b = map(int, input().split())
    for i in range(a, b + 1) :
        for j in range(i + 1, b + 1) :
            if (gcd(i, j) == 1) :
                print('(', i, ',', j, ')', sep = '')
```
